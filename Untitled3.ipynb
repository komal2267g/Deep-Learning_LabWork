```python
import numpy as np

# sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))


# given values

x=np.array([1,0])

y=1


W1=np.array([[0.2,0.3],
             [0.4,0.1]])

W2=np.array([[0.5],
             [0.6]])


# forward propagation

hidden_input=np.dot(x,W1)

hidden_output=sigmoid(hidden_input)

output_input=np.dot(hidden_output,W2)

prediction=sigmoid(output_input)

print("Hidden Layer Activations")
print(hidden_output)

print("\nOutput Prediction")
print(prediction)


# loss

loss=(y-prediction)**2

print("\nLoss")
print(loss)


# backpropagation

d_loss=2*(prediction-y)

delta_output=d_loss*(prediction*(1-prediction))

print("\nOutput Delta")
print(delta_output)


# gradients for W2

grad_W2=hidden_output.reshape(-1,1)*delta_output

print("\nGradient of W2")
print(grad_W2)


# hidden layer gradients

hidden_delta=(delta_output*W2.T)*(hidden_output*(1-hidden_output))

grad_W1=np.outer(x,hidden_delta)

print("\nGradient of W1")
print(grad_W1)


# learning rate

lr=0.1


# updated weights

new_W1=W1-(lr*grad_W1)

new_W2=W2-(lr*grad_W2)

print("\nUpdated W1")
print(new_W1)

print("\nUpdated W2")
print(new_W2)
```
