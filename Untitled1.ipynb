```python
import numpy as np

# XOR Dataset

X=np.array([[0,0],
            [0,1],
            [1,0],
            [1,1]])

y=np.array([[0],
            [1],
            [1],
            [0]])

print("Input Data:\n")
print(X)

print("\nOutput Labels:\n")
print(y)
```

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt
```

```python
model_mse=Sequential()

model_mse.add(

    Dense(
        4,
        activation="sigmoid",
        input_shape=(2,)
    )

)

model_mse.add(

    Dense(
        1,
        activation="sigmoid"
    )

)


model_mse.compile(

    optimizer="adam",

    loss="mse",

    metrics=["accuracy"]

)


history_mse=model_mse.fit(

    X,
    y,

    epochs=500,

    verbose=0

)
```

```python
model_bce=Sequential()

model_bce.add(

    Dense(
        4,
        activation="sigmoid",
        input_shape=(2,)
    )

)

model_bce.add(

    Dense(
        1,
        activation="sigmoid"
    )

)


model_bce.compile(

    optimizer="adam",

    loss="binary_crossentropy",

    metrics=["accuracy"]

)


history_bce=model_bce.fit(

    X,
    y,

    epochs=500,

    verbose=0

)
```

```python
print("MSE MODEL\n")

mse_loss,mse_accuracy=model_mse.evaluate(
    X,
    y,
    verbose=0
)

print("Loss =",mse_loss)
print("Accuracy =",mse_accuracy)


print("\n----------------------\n")


print("BCE MODEL\n")

bce_loss,bce_accuracy=model_bce.evaluate(
    X,
    y,
    verbose=0
)

print("Loss =",bce_loss)
print("Accuracy =",bce_accuracy)
```

```python
print("MSE Predictions\n")

print(

model_mse.predict(X)

)



print("\n-----------------\n")


print("BCE Predictions\n")

print(

model_bce.predict(X)

)
```

```python
plt.figure(figsize=(8,5))

plt.plot(

history_mse.history["loss"],
label="MSE"

)

plt.plot(

history_bce.history["loss"],
label="BCE"

)

plt.xlabel("Epochs")

plt.ylabel("Loss")

plt.title("MSE vs BCE")

plt.legend()

plt.grid(True)

plt.show()
```
