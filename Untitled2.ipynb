```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
```

```python
# creating a binary classification dataset

X,y=make_classification(

    n_samples=300,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    class_sep=1.2,
    flip_y=0.05,
    random_state=42

)

# reshape labels

y=y.reshape(-1,1)

# normalize dataset

X=(X-X.mean(axis=0))/X.std(axis=0)


# splitting dataset

X_train,X_test,y_train,y_test=train_test_split(

    X,
    y,
    test_size=0.25,
    random_state=42

)

print("Training Samples =",len(X_train))
print("Testing Samples =",len(X_test))
```

```python
model_mse=Sequential()

model_mse.add(
    Input(shape=(2,))
)

model_mse.add(

    Dense(
        8,
        activation="sigmoid"
    )
)

model_mse.add(

    Dense(
        1,
        activation="sigmoid"
    )
)


model_mse.compile(

    optimizer="adam",

    loss="mse",

    metrics=["accuracy"]

)


history_mse=model_mse.fit(

    X_train,
    y_train,

    epochs=300,

    validation_data=(X_test,y_test),

    verbose=0

)
```

```python
model_bce=Sequential()

model_bce.add(

    Input(shape=(2,))
)

model_bce.add(

    Dense(
        8,
        activation="sigmoid"
    )
)

model_bce.add(

    Dense(
        1,
        activation="sigmoid"
    )
)


model_bce.compile(

    optimizer="adam",

    loss="binary_crossentropy",

    metrics=["accuracy"]

)


history_bce=model_bce.fit(

    X_train,
    y_train,

    epochs=300,

    validation_data=(X_test,y_test),

    verbose=0

)
```

```python
print("\nMODEL-A (MSE)\n")


mse_loss,mse_accuracy=model_mse.evaluate(

    X_test,
    y_test,
    verbose=0

)


print("Loss =",mse_loss)
print("Accuracy =",mse_accuracy)


print("\n----------------------------------")


print("\nMODEL-B (BCE)\n")


bce_loss,bce_accuracy=model_bce.evaluate(

    X_test,
    y_test,
    verbose=0

)


print("Loss =",bce_loss)
print("Accuracy =",bce_accuracy)
```

```python
plt.figure(figsize=(8,5))


plt.plot(

    history_mse.history["loss"],
    label="MSE"

)

plt.plot(

    history_bce.history["loss"],
    label="BCE"

)


plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss Comparison")

plt.legend()

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))


plt.plot(

    history_mse.history["accuracy"],
    label="MSE"

)

plt.plot(

    history_bce.history["accuracy"],
    label="BCE"

)


plt.xlabel("Epochs")
plt.ylabel("Accuracy")

plt.title("Training Accuracy Comparison")

plt.legend()

plt.grid(True)

plt.show()
```

```python
print("Final Training Accuracy\n")


print(

"MSE =",
history_mse.history["accuracy"][-1]

)

print(

"BCE =",
history_bce.history["accuracy"][-1]

)


print("\nFinal Training Loss\n")


print(

"MSE =",
history_mse.history["loss"][-1]

)


print(

"BCE =",
history_bce.history["loss"][-1]

)
```
